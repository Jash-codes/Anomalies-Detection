```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Jashwanth\OneDrive\Documents\Anomalies Detection\Dataset\stock_market_dataset_50000_rows.csv')
df_100 = df.head(100)
df.columns = df.columns.str.strip()
```

```python
df_100 = df.head(100).copy()

df_100.loc[:, 'Timestamp'] = pd.to_datetime(df_100['Timestamp'])
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_100['Timestamp'], df_100['Price'], color='blue', label='Stock Price')
plt.title("Stock Price Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
top_volume = df_100.groupby('Company')['Volume'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_volume.index, y=top_volume.values, palette='viridis')
plt.title("Top 10 Companies by Total Volume")
plt.ylabel("Volume")
plt.xlabel("Company")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_100, x='Price', y='Volume', hue='Company', palette='Set2', legend=False)
plt.title("Price vs Volume")
plt.xlabel("Price")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_100, x='Company', y='Price', palette='pastel')
plt.title("Price Distribution per Company")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_100['%Change'], bins=20, kde=True, color='coral')
plt.title("Distribution of % Change")
plt.xlabel("% Change")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
